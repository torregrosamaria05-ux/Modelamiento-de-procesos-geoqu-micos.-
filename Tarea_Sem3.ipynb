EJERCICIO 1:

```python
precip_months = ['jan' , 'feb' , 'mar' , 'apr' , 'may' , 'jun' , 'jul', 'aug' , 'sep' , 'oct' , 'nov' , 'dec' ] #Creé la lista de meses
```

```python
precip_inches = [0.70 , 0.75 , 1.82 , 2.93 , 3.05 , 2.02 , 1.93 , 1.62 , 1.84 , 1.31 , 1.39 , 0.84 ] #Creé la lista de pulgadas
```

EJERCICIO 2:

```python
inch_to_mm =  25.4
```

```python
precip_mm = precip_inches
precip_mm = [x * inch_to_mm for x in precip_mm] #Aquí convertí las variables de pulgadas a milimetros
```

EJERCICIO 3:

```python
all_precip_data = [precip_months , precip_mm] #Creé la lista de listas
print(all_precip_data)
```

PREGUNTAS REFLEXIVAS:
- Agrupar datos relacionados en una sola estructura es muy útil porque mantiene todo organizado, haciendo el código más fácil de entender y de trabajar. Así, nos aseguramos de que la información esté coherente, evitamos errores al manipularla y simplificamos mucho si tenemos que pasar esos datos a otra parte del programa.
- Una lista de listas, a diferencia de una tabla como un DataFrame, presenta varias limitaciones clave. Le falta la claridad de los nombres de columna explícitos, lo que complica el acceso y la comprensión de los datos y es ineficiente y propensa a errores para operaciones complejas como filtrar, ordenar o realizar análisis estadísticos, especialmente con grandes volúmenes de información, donde una tabla ofrece mejor estructura

EJERCICIO 4:

```python
# Importar la librería necesaria
import matplotlib.pyplot as plt

# Crear figura y ejes
fig, ax = plt.subplots(figsize=(8, 5))

# Graficar datos (modifica estas líneas)
ax.bar(
    all_precip_data[0],
    all_precip_data[1],
    color="turquoise"
)

# Configurar títulos y etiquetas
ax.set(
    title="Precipitación anual",
    xlabel="Tiempo (meses)",
    ylabel="Precipitación (mm)"
)

plt.show()
```

PREGUNTAS DE REFLEXIÓN:
- Es fundamental entender cómo están organizados los datos antes de graficar porque la estructura impacta directamente la forma en que los visualizamos, saber qué datos corresponden a los ejes, las barras o los puntos nos permite elegir el tipo de gráfico correcto y asegurar que la representación sea precisa, legible y transmita el mensaje deseado sin confusiones.
- Como se mencionó en las respuestas anteriores las listas carecen de estructura, nombres de columnas explícitos y métodos eficientes para análisis de datos complejos , lo que sí ofrece un DataFrame, que está optimizado para estas operaciones y grandes volúmenes de datos.
